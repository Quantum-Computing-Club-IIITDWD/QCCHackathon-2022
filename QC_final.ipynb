```python
from qiskit import *
%matplotlib inline
```

```python
qc=QuantumCircuit(3,3)
```

```python
qc.draw(output ='mpl')
```

```python
qc.y(0)
qc.h(0)
qc.y(0)
qc.draw(output ='mpl')
```

```python
qc.barrier(0,1,2)
qc.h(1)
qc.cx(1,2)
qc.barrier(0,1,2)
```

```python
qc.draw(output ='mpl')
```

```python
qc.h(1)
qc.barrier(0,1,2)
qc.draw(output ='mpl')
```

```python
qc.measure(0,0)
qc.measure(1,1)
qc.barrier(0,1,2)
```

```python
qc.draw(output ='mpl')
```

```python
qc.cx(1,2)
qc.cz(0,2)
qc.y(2)
qc.h(2)
qc.y(2)
qc.measure(2,2)
qc.draw(output ='mpl')
```

```python
from qiskit.tools.visualization import *
backend=Aer.get_backend('aer_simulator')
result=backend.run(qc).result()
count=result.get_counts()
plot_histogram(count)
```
